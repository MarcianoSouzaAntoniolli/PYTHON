```python
# operadores boleano verdadeiro ou falso
valor1 = input("Informe o valor 1: ")
valor2 = input("Informe o valor 2: ")
print(f"resutado de maior          {valor1} >  {valor2} ? {valor1 > valor2}")
print(f"resutado de menor          {valor1} <  {valor2} ? {valor1 < valor2}")
print(f"resutado de maior ou igual {valor1} >= {valor2} ? {valor1 >= valor2}")
print(f"resutado de menor ou igual {valor1} <= {valor2} ? {valor1 <= valor2}")
print(f"resutado de difetente      {valor1} != {valor2} ? {valor1 != valor2}")
print(f"resutado de igualdade      {valor1}  == {valor2}? {valor1 == valor2}")

# >
# <
# >=
# <=
# != diferente
# == iguarde
```

```python
#programa que calcule o o imc de um atleta
#entra
try:
  altura = float(input("Informe sua Altura: "))
  peso   = float(input("Informe seu peso: "))
except:
  print("OCORREU UM ERRO TENTE NOVAMENTE")


#processamento
imc = peso / (altura*altura);
print(f"Resultado do deu IMC é: {round(imc,2)} ")

#saida
print(f"Resultado do deu IMC é: {round(imc,2)} ")
if imc < 18.5: # if :
  print ("****BAIXO PESO****")
else:
   if imc > 18.5 and imc < 24.9:
    print ("****PESO NORMAL****")
   else:
    print ("****SOBREPESO****")


print ("")
print ("")
print("Abaixo de 18,5: baixo peso")
print("Entre 18,5 e 24,9: peso normal")
print("Entre 25 e 29,9: sobrepeso")
```
